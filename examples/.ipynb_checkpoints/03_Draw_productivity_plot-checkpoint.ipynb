```python
import sys
sys.path.append("../.")
import warnings
warnings.filterwarnings('ignore')

import pickle
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from matplotlib import cm
from ast import literal_eval
from tqdm import tqdm
from functools import partial
from sklearn.preprocessing import MinMaxScaler

from dstar.apps import calculator, mask
```

```python
# Load scaler and coordination number

with open('./utils/scaler.pkl','rb') as fr:
    scaler_dict = pickle.load(fr)
with open('./utils/cn.pkl','rb') as fr:
    cn_dict = pickle.load(fr)
```

```python
eU_lst = np.arange(-1.6,-0.7,0.05) # Range of applied potential to calculate producitivty
eU_dict = {}

path = './energy/' # Path for dataframes with predicted binding energies
csv_lst = os.listdir(path)

# Collect productivities by applied potential

for eU in tqdm(eU_lst):
    csv = 'Ag_Cu.csv' # Target combination

    df = pd.read_csv(path+csv,index_col=0)
    df.drop_duplicates(['FNN','Same','Sub'],inplace=True)
    csv = csv.split('.')[0]
#     masked_df = masking.masking_dataframe(df, fnn='Cu',host='Cu',composition=7,comp_operator='>',cn=7, cn_operator='<') # Use to draw masked productivity plot
    masked_df = df
    
    # Predicted binding energies
    co_arr = masked_df['CO_energy'].to_numpy()
    h_arr = masked_df['H_energy'].to_numpy()
    oh_arr = masked_df['OH_energy'].to_numpy()
    
    g_max_by_product= calculator.get_activity(co_arr, h_arr, oh_arr, eU) # Calculate raw productivity from the delta_G_max
    total_num = len(co_arr)
    productivity = calculator.boltzman_product(g_max_by_product,total_num)
    
    for i in range(4):
        productivity[i] = scaler_dict[eU][i].transform(np.array([productivity[i]]).reshape(-1,1))[0][0]
    
    eU_dict[eU] = productivity
```

```python
# Draw productivity by potential

blue = '#6661AB'
red = '#C8566B'
purple = '#9D75BF'
green = '#9EC299'

color_lst = [green,red,blue,purple]
label_lst = ['HCOOH', 'CO', 'C1', 'HER']
fig, ax = plt.subplots(1,1,figsize=(10,8))
for i in [0,1,2,3]:
    prod_lst = [eU_dict[eU][i] for eU in eU_lst]
    ax.plot(eU_lst, prod_lst, lw = 2, marker='o',markersize=10,color=color_lst[i], label=label_lst[i])
ax.legend(fontsize=20)
ax.set_ylim([-0.05,1.05])
ax.set_xlim(-1.635, -0.765)
[ax.spines[i].set_linewidth(2) for i in ax.spines.keys()]
ax.tick_params(width = 2, size = 7, labelsize = 20)
ax.set_ylabel('Activity',fontsize=25)
ax.set_xlabel('U V vs RHE',fontsize=25)
plt.savefig('./figure/potential.png')
```
